```python
# print the state space and action space
print(env.observation_space)
print(env.action_space)

# print the total number of states and actions
print(env.nS)
print(env.nA)
```

```python
# import the dynamic programming algorithms
import sys
sys.path.append('../')
from algorithms.dp import *
```

```python
# define policy
random_policy = np.ones([env.nS, env.nA]) / env.nA
```

```python
V = policy_evaluation(env, random_policy, gamma=0.95)
plot_values(V)
```

```python
# obtain the optimal policy and optimal state-value function
policy_pi, V_pi = policy_iteration(env, gamma=0.95)

# print the optimal policy
print("\nOptimal Policy (LEFT = 0, DOWN = 1, RIGHT = 2, UP = 3):")
print(policy_pi,"\n")

plot_values(V_pi)
```

```python
policy_vi, V_vi = value_iteration(env, gamma=0.95, epsilon=1e-8)

# print the optimal policy
print("\nOptimal Policy (LEFT = 0, DOWN = 1, RIGHT = 2, UP = 3):")
print(policy_vi,"\n")

# plot the optimal state-value function
plot_values(V_vi)
```
